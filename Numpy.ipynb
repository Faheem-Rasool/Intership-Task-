Indexing & slicing



```python
a=[1,3,4,5,2]
result_indices = np.where(np.array(a) == 2)
if result_indices[0].size > 0:
    print("Found")
```

```python
a=[1,3,4,5,2]
arr=np.take(a,2)
print(arr)
```

```python
a=[1,3,4,5,2]
arr=np.unique(a)
print(arr)
```

```python
a=[1,3,4,5,2]
arr=np.nonzero(a)
print(arr)
```

```python
a=[1,3,4,5,2]
arr=np.argsort(a)
print(arr)
```

```python
a=np.array([1,3,4,5,2])
arr=np.split(a,[2])
print(arr)
```

```python
a=[1,3,4,5,2]
arr=np.flip(a)
print(arr)
```

```python
a=[1,3,5,2]
arr=np.roll(a, shift=0)
print(arr)
```
