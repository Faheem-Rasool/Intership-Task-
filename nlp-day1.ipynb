```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
test_food="I am now officially Engineer by PEC"
words=word_tokenize(test_food)
print(f"Word_token:{words}")
```

```python
text_chat="hi! where are you? I am at home."
sentence=sent_tokenize(text_chat)
print(f"Sentence Tokens are:{sentence}")
```

```python
import re
from nltk.corpus import stopwords
```

```python
text_review="This is a very good product and I like it"
stop_words=set(stopwords.words("english"))
words=word_tokenize(text_review)
print(stop_words)
print(words)
```

```python
clean_review=[w for w in words if w.lower() not in stop_words]
print(f"Cleaned Review:{clean_review}") 
```

```python
text_social="Wow!!!!!! This @phone is amazing!!!!!"
clean_social=re.sub(r"[^\w\s]","",text_social)
print(f"cleaned Social Post:{clean_social}")
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
word_fit=["fascinatingly","studies","intensities","running","walking","exercising"]
stem_fit=[stemmer.stem(w) for w in word_fit]
print(f"Fitness Stem:{stem_fit}")
```

```python
email_text = "Hello sir! please check my portfolio"
print(f"normalize case:{email_text.upper()}")
print(f"normalize case:{email_text.lower()}")
```

```python
chat_text_A = "U r Amazing"
normalized_text = chat_text_A.replace("U", "You").replace("r", "are")
print(f"Corrected Chat: {normalized_text}")
```

```python
import re
text_phone_1="Call me at +923408003646"
numbers=re.findall(r"\d+",text_phone_1)
print(f"Phone Numbers:{numbers}")
```

```python
text_email="My email is uajkfaheem@gmail.com.You can submit your assignment here"
email=re.findall(r"\S+@\S+",text_email)
print(f"Emals found:{email}")
```

```python
text_tag="#ILoveCoding"
words_tag=re.findall("[A-Z][a-z]*",text_tag)
print(f"segmented words:{words_tag}")
```
