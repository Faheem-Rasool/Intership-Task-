```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc 
```

```python
features = ['driver_age', 'driver_gender', 'driver_race', 'search_conducted', 'drugs_related_stop']
```

```python
features = ['driver_age', 'driver_gender', 'driver_race', 'search_conducted', 'drugs_related_stop']
if set(features + ["is_arrested"]).issubset(df.columns):
    model_df = df[features + ['is_arrested']].dropna()
    print(f"Model Daata Frame Created . Size : {model_df.shape[0]}")
else:
    raise Exception("Missing required columns for modeling")
```

```python
df.columns
```

```python
df.info()
```

```python
model_df = pd.get_dummies(model_df, columns=["driver_gender", "driver_race"], drop_first=True)
bool_cols = ['search_conducted' , 'drugs_related_stop']
for col in bool_cols:
    if model_df[col].dtype == bool:
        model_df[col]= model_df[col].astype(int)
print("Finishing converting the text into numbders")
```

```python
X = model_df.drop("is_arrested", axis= 1 )
Y= model_df["is_arrested"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
print(f"Training set size:,{X_train.shape[0]}  rows")
```

```python
print(f"Testing set size:,{X_test.shape[0]}  rows")
```

```python
try:
    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train, Y_train)
    y_prediction = clf.predict(X_test)
    print("Model trained successfully")
    accuracy = accuracy_score(Y_test, y_prediction)
    print(f"Final Grade (Accuracy): {accuracy:.2f} 0r {accuracy*100:.1f}%")
except Exception as e:
    print("Error training the model: ", e)
```

```python
conf_matrix = confusion_matrix(Y_test, y_prediction)
plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot= True, fmt= "d", cmap = "Blues")
plt.title("Confusion Matrix")
plt.xlabel("What Model Predicted?")
plt.ylabel("What Actually Happened?")
plt.tight_layout()
plt.show()
```

```python
y_props = clf.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(Y_test, y_props)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label= f"ROC curve (Ae("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc= "lower right")
plt.tight_layout()
plt.show()
```
