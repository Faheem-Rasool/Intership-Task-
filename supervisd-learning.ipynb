```python
import pandas as pd
```

```python
from sklearn import linear_model
```

```python
import matplotlib.pyplot as plt
```

```python
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
data={
    "Area":[2600,3000,3200,3600,4000],
    "price":[55000,565000,610000,680000,725000]

}
df=pd.DataFrame(data)
```

```python
df.to_csv("Home Prices.csv",index=False)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df.Area,df.price,color="Blue",marker="+",s=100)
plt.title("Label Data")
plt.xlabel("Area(sq ft)")
plt.ylabel("price(USD)")
plt.grid(True)
plt.show()
```

```python
bad_m=10
bad_b=200000
X_line=np.array([2000,4500])
y_bad_guess=bad_m*X_line+bad_b
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df.Area,df.price,color="Blue",marker="+",s=100)
plt.plot(x_line, y_bad_guess,color="purple",linestyle="--",label="Untrained hypothesis")
plt.title("Untrained hypothesis(Guess VS Truth")
plt.xlabel("Area(sq ft)")
plt.ylabel("price(USD)")
plt.grid(True)
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["Area"]],df.price)
```

```python
learned_m=reg.coef_[0]
learned_b=reg.intercept_
y_learned_fit=learned_m*X_line+learned_b
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df.Area,df.price,color="Blue",marker="+",s=100)
plt.plot(X_line, y_learned_fit,color="green",linewidth=2.5,label="optimized hypothesis")
plt.title("Untrained hypothesis(Guess VS Truth")
plt.xlabel("Area(sq ft)")
plt.ylabel("price(USD)")
plt.grid(True)
plt.show()
```

```python
test_area =3300
prediction=reg.predict([[test_area]])
print(f"Pridiction For{test_area}sq ft:${prediction}")
```
