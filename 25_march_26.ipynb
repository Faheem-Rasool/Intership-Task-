```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.boxplot(data=df,x="Weight (kg)")
plt.show()
```

```python
sns.boxplot(data=df,x="Standing Height (cm)")
plt.show()
```

```python
sns.displot(data=df,x="Weight (kg)",kind="hist")
plt.show()
```

```python
sns.displot(data=df,x="Standing Height (cm)",kind="hist")
plt.show()
```

```python
sns.scatterplot(data=df,x="Weight (kg)",y="Standing Height (cm)")
plt.show()
```

```python
df.skew()
```

```python
z_score_height = stats.zscore(df["Standing Height (cm)"])
print(z_score_height)
```

```python
df1=df.copy()
```

```python
df1["Z score Height"]=z_score_height
```

```python
df1
```

```python
z_score_height = stats.zscore(df["Weight (kg)"])
print(z_score_height)
```

```python
df1["Z score Weight"]=z_score_height
```

```python
df1
```

```python
plt.figure(figsize =(16,5))
plt.subplot(1,2,1)
sns.histplot(df1["Standing Height (cm)"])

plt.subplot(1,2,2)
sns.histplot(df1["Z score Height"])
plt.show()
```

```python
plt.figure(figsize =(16,5))
plt.subplot(1,2,1)
sns.histplot(df1["Weight (kg)"])

plt.subplot(1,2,2)
sns.histplot(df1["Z score Weight"])
plt.show()
```
