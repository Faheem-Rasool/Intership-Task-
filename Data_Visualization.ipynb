```python
import pandas as pd
Data=pd.read_csv("/content/data.csv") # Reload Data to ensure it's not None
Data = Data.drop (["Engine Fuel Type","Market Category",
                  "Number of Doors","Vehicle Style",
                  "Popularity","Vehicle Size"],axis = 1)
Data.head()
```

```python
Data.info()
```

```python
Data=Data.rename(columns ={"Engine HP": "HP","Engine Cylinders":"Cylinders",
                          "Driven_Wheels":"Drive Mode", "highway MPG":"HPG-H"})
Data.head()
```

```python
Data.shape
```

```python
Duplication=Data[Data.duplicated()]
print("Numbers of Duplicates Rows",  Duplication.shape)
```

```python
Data.count()
```

```python
Data=Data.drop_duplicates()
Data.count()
```

```python
print(Data.isnull().sum())
```

```python
Data=Data.dropna()
```
